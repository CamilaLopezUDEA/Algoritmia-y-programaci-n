Clasificación de los nombres para la guardar cada uno en una lista

```python
#Función que remueve el salto de línea
def clearBack(l1):
    return [i[1:-1] for i in l1]

#Función elimina los nombres y apellidos que están repetidos
def eliminarRepetidos(lista):
    lista.sort()  #Función que ordena los elementos por order alfabético en este caso
    repetidos = []

    for i in range(len(lista)):#En estos ciclos se guardan todos los elementos que se repitan en la lista 'repetidos'
        contador = 0
        for e in range(len(lista)):
            if lista[i] == lista[e]:
                contador += 1
                if contador == 2:
                    repetidos.append(lista[i])
        contador = 0

    for i in range(len(repetidos)):
        lista.remove(repetidos[i])  #Removemos de la lista final todos los elementos que se encuentran repetidos
        if i < len(repetidos) - 1:
            if repetidos[i] != repetidos[i + 1]:   #Volvemos a añadir sólo un ejemplar del elemento que está repetido a la lista final
                lista.append(repetidos[i])

    return lista

#Función que clasifica los nombres por género masculino y femenino
def organizarPorGenero(nombres):
    excepcionesFem = ["Emily", "Isabel", "Jennifer", "María José"]  #Nombres femeninos que no terminan con "a"
    excepcionesFfem2 = ['Antonella', 'Laura', 'Lucía', 'Sofia', 'Victoria', 'Catalina', 'Isabella'] #Nombres femeninos omitidos por el primer ciclo de la función
    masculinos = nombres
    femeninos = []

    for nombre in nombres:         #Removemos los nombres femeninos que terminan con "a" de la lista de nombres masculinos
        if nombre[-1] == "a":
            masculinos.remove(nombre)
            femeninos.append(nombre)
        if nombre[-1] == "Laura":
            print(nombre)

    for nombre in excepcionesFem:  #Removemos los nombres femeninos de la lista de nombres masculinos
        masculinos.remove(nombre)
        femeninos.append(nombre)

    for nombre in excepcionesFfem2:  #Removemos los nombres femeninos de la lista de nombres masculinos nuevamente
        masculinos.remove(nombre)
        femeninos.append(nombre)

    return (femeninos, masculinos)  #Retornamos tupla cuyo primer elemento son los nombres femeninos y el segundo los femeninos

#Función que extrae los nombres de los archivos y los organiza
def organizarListas():
    archivo = open("/content/drive/MyDrive/Algoritmia y Programación/Trabajo Final - Camila Lopez Cordero/nombres.csv")
    nombres = clearBack(archivo.readlines()[1:])  #Lista donde se encuentran los nombres
    archivo = open("/content/drive/MyDrive/Algoritmia y Programación/Trabajo Final - Camila Lopez Cordero/apellidos.csv")
    apellidos = clearBack(archivo.readlines()[1:])  #Lista donde se encuentran los apellidos

    nombres = eliminarRepetidos(nombres)    #Llamamos función para eliminar nombres repetidos
    apellidos = eliminarRepetidos(apellidos)

    nombresFinales = organizarPorGenero(nombres)  #Llamamos función para separar nombres masculinos de femeninos

    return (nombresFinales[0], nombresFinales[1], apellidos)  #Retornamos tupla cuyo primer elemento son los nombres femeninos, el segundo los femeninos y el tercero, los apellidos

```

Generación de los nombres completos finales

```python
#Función para generar un nombre compuesto de 2 palabras
def generarNombre(nFem, nMas):        #Parámetro = Lista de nombres

     nList = []

    #Este Bloque determina si el nombre generado es masculino o femenino. Si la variable género es par, es femenino
     genero = rint(1, 20)
     if genero % 2 == 0:
         nList = nFem
     else:
        nList = nMas

     i = rint(0, (len(nList) - 1))
     e = i

     #Siempre que la posición de ambos nombres en la lista sea la misma, el ciclo continuará ejecutándose
     while i == e:
        e = rint(0, (len(nList) - 1))

     return nList[i] + " " + nList[e]

#Función para reorganizar la posicion de los elementos en una lista aleatoriamente
def reorganizarLista(lista):    #Parámetro = lista
    randomList = []
    while len(lista) > 1:

        if len(lista) == 2:
            randomList.append(lista[0])
            randomList.append(lista[1])

            return randomList

        i = rint(0, (len(lista) - 1))
        randomList.append(lista[i])
        lista.remove(lista[i])

#Función que crea "num" combinaciones de 2 nombres y 2 apellidos
def crearNombresCompletos(nFem, nMas, a, num):  #Parámetros = Nombres femeninos, Nombres masculinos, Apellidos, Número de nombres a crear
    nombresCompletos = []
    i = 0

    posicion = rint(0,(len(a) - 2))

    #En este ciclo se generan "num" combinaciones de nombres y apellidos que nunca se repiten
    while i < num:
        posicion2 = posicion + 1
        while posicion2 != posicion:
            #Añade el nombre completo a la lista
            nombresCompletos.append(generarNombre(nFem, nMas) + " " + a[posicion] + " " + a[posicion2])
            posicion2 += 1

            if posicion2 == len(a):
               posicion2 = 0

            i += 1
            if i == num:
                posicion2 = posicion

        posicion += 1
        if posicion == len(a):
            posicion = 0

    return nombresCompletos
```

```python
#Guardamos cada elemento en una lista final
lista = organizarListas()

nFemeninos = lista[0]
nMasculinos = lista[1]
apellidos = lista[2]

#Llamamos las funciones para generar la lista de nombres completos ya para reordenar esta lista
nombresCompletos = reorganizarLista(crearNombresCompletos(nFemeninos, nMasculinos, apellidos, 1000))
```

```python
#En este caso crearemos una función que remueva los espacios de una cadena de texto, esto con el propósito de crear la lista de nombres capitalizados para todas las materias
def quitarEspacios(cadena):
    cadena = cadena.replace(" ", "")
    return cadena

#Ahora creamos un dataframe con todos los elementos del archivo con la información de cada materia
df_infoMaterias = pd.read_csv("/content/drive/MyDrive/Algoritmia y Programación/Trabajo Final - Camila Lopez Cordero/ListaMaterias.csv")

#Removemos los espacios de la columna de
df_infoMaterias["Nombre"] = [quitarEspacios(nombre) for nombre in df_infoMaterias["Nombre"]]
df_infoMaterias
```

Ahora comenzamos a hacer la planeación para el siguiente semestre

```python
#Para el Código Asignatura(CA)

#Primero calculamos el número de grupos para cada materia de X semestre
porcentajeProporcional = [i for i in reversed(range(6, 15))]
porcentajeProporcional.insert(4, 10)

cuposAula = [30, 30, 30, 25, 25, 25, 20, 20, 20, 10]
cantidadGrupos= []

for i in range(0, 10):
    if (porcentajeProporcional[i] * 10) % cuposAula[i] == 0:
        cantidadGrupos.append(int(porcentajeProporcional[i] * 10 / cuposAula[i]))
    else:
        cantidadGrupos.append(int(porcentajeProporcional[i] * 10 / cuposAula[i]) + 1)


CA = []

for codigo in df_infoMaterias["Código (sin grupo)"][:-1]:
    #En este caso a la variable semestre le restamos una unidad ya que las listas de python empiezan desde la posición 0
    semestre = int(codigo[3]) - 1

    for i in range(0, cantidadGrupos[semestre]):
        CA.append(codigo + str(i + 1))

#Como en el 10mo semestre el código es diferente, añade el código de la única materia del semestre 10, Práctica profesional > P101210
CA.append(df_infoMaterias["Código (sin grupo)"].iloc[-1] + "10")
```

```python
#Calculamos Horas de trabajo docente (HTD)
HTD = []

for codigo in CA[:-1]:
    #En este caso a la variable semestre le restamos una unidad ya que las listas de python empiezan desde la posición 0
    semestre = int(codigo[4])

    if semestre == 1:
        HTD.append(16)
    elif semestre == 2:
        HTD.append(32)
    elif semestre == 3:
        HTD.append(64)
    elif semestre == 4:
        HTD.append(96)

#Como en el 10mo semestre no hay horas de trabajo docente, añadimos 0 al final para la única materia de este semestre
HTD.append(0)
```

```python
#Calculamos Horas de trabajo independiente (HTI)
HTI = []

for codigo in CA[:-1]:
    #En este caso a la variable semestre le restamos una unidad ya que las listas de python empiezan desde la posición 0
    semestre = int(codigo[4])

    if semestre == 1:
        HTI.append(32)
    elif semestre == 2:
        HTI.append(64)
    elif semestre == 3:
        HTI.append(80)
    elif semestre == 4:
        HTI.append(120)

#Como en el 10mo semestre no hay horas de trabajo independiente, añadimos 0 al final para la única materia de este semestre
HTI.append(0)
```

```python
#Calculamos Número total de estudiantes (NTE)
NTE = []
for grupo in CA:

    #En este condicional añadimos el número total de estudiantes para práctica profesional manualmente
    if grupo[:5] == "P1012":
        NTE.append(60)
        break

    NTE.append(porcentajeProporcional[int(grupo[3]) - 1] * 10)
```

```python
#Calculamos el código del curso (CC)
CC = [int(grupo[-1]) for grupo in CA]
```

```python
#Calculamos el Total de Cursos Asignados (TCA)
TCA = []
for i in range(0, len(CA) - 1):
    TCA.append(cantidadGrupos[int(CA[i][3]) - 1])

#Añadimos el total de cursos asignados para la práctiac profesional
TCA.append(6)
```

```python
#Calculamos la fecha de creación para cada grupo (FC)
FC = [date.today() for i in range(len(CA))]
```
